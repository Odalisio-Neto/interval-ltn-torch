```python
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon
import matplotlib.colors as mcolors
import numpy as np
from skspatial.measurement import area_signed

import iltn
from iltn.relations.trapz.intersection import find_intersection_vertices, shoelace_formula
```

```python
x = iltn.events.TrapzEvent("A", [0.5,1.5,3,5], trainable=False)
y = iltn.events.TrapzEvent("B", [3.,4,5.,6.], trainable=True)
vertices = find_intersection_vertices(x,y)
vertices_array = np.array([[v.x,v.y] for v in vertices])

fig, ax = plt.subplots(1,1,figsize=iltn.utils.plot.set_size(300., subplots=(1, 1)))
iltn.utils.plot.plot_events(np.arange(0.,10.,0.1), [x,y], ax)
ax.scatter([v.x for v in vertices], [v.y for v in vertices], color="green")
for (i,v) in enumerate(vertices):
    ax.annotate(i, (v.x,v.y))
patch = Polygon(vertices_array, fc=mcolors.to_rgba("green",alpha=0.2))
ax.add_patch(patch)
ax.text(.01, .99, f"Area={shoelace_formula(vertices):.2f}", 
        ha='left', va='top', weight='bold', transform=ax.transAxes)
print(area_signed(vertices_array))
ax.legend()
ax.plot()
```
